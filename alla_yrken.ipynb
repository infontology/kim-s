# Alla yrken från Arbetsförmedlingen

```python
import requests
```

Arbetsförmedlingen har ett API för att kunna få ut alla yrken, med eller utan beskrivningar. Här tar vi ut listan i textformat, och avslutar med att ta fram detaljer för ett yrke utifrån ID.

Dokumentation: http://api.arbetsformedlingen.se/af/v0/Occupation/wsoccupation.asmx?op=GetAllOccupations 

Python-hjälp: https://stackoverflow.com/questions/18175489/sending-soap-request-using-python-requests 

```python
url="http://api.arbetsformedlingen.se/af/v0/Occupation/wsoccupation.asmx"
#headers = {'content-type': 'application/soap+xml'}
headers = {'content-type': 'text/xml'}
body = """<?xml version="1.0" encoding="utf-8"?>
<soap:Envelope xmlns:xsi="http://www.w3.org/2001/XMLSchema-instance" xmlns:xsd="http://www.w3.org/2001/XMLSchema" xmlns:soap="http://schemas.xmlsoap.org/soap/envelope/">
  <soap:Body>
    <GetAllOccupations xmlns="urn:ams.se:wsoccupation" />
  </soap:Body>
</soap:Envelope>"""

#GetAllOccupations har varianterna GetAllOccupationsDetailed och GetAllOccupationsShort
```

```python
response = requests.post(url,data=body,headers=headers)
#response.content #Om man vill få ut rå xml
```

```python
import xml.etree.ElementTree as ET 
tree = response.content.decode('utf-8')
root= ET.fromstring(tree)
```

Jag tror att det ställer till det vid parsningen att det finns en massa pekare till name spaces, men jag vet för lite om xml egentligen...

```python
root
```

Yrkena vi vill iterera över ligger tre steg ner i hierarkin:

```python
occupations = root[0][0][0]
```

```python
yrken=[]
yrke = []
for occupation in occupations:
    print(occupation[0].text, occupation[1].text)
    yrke = [occupation[0].text,occupation[1].text]
    yrken.append(yrke)
```
